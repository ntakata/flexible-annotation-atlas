# Merge AO_L.json and AO_R.json
This note book combines two json files (AO_L.json and AO_R.json) to prepare a json file (AO_LR_wo_VC.json) that assigns different IDs for homotopic area in right and left sides of the brain.

# Set variables

```python
dir_data = 'data'

fn_input_AOL = 'AO_L.json'
fn_input_AOR = 'AO_R.json'
fn_input_template = 'AO_LR_wo_TEMPLATE.json'

fn_output_AO = 'AO_LR_wo_VC.json'
```

```python
import os
import json
from collections import OrderedDict
```

# Merge jsons

```python
def delete_recur(node, target):
    if node["acronym"] in target:
        print("ERROR")
    children = node["children"]
    
    if children == None:
        return
    
    new_children = list()
    for child in children:
        if child["acronym"] not in target:
            new_children.append(child)
        else:
            print("DELETED {}".format(child["acronym"]))
    for child in new_children:
        delete_recur(child, target)
        
    node["children"] = new_children
```

```python
with open(os.path.join(dir_data, fn_input_AOL), "r") as f:
    aol = json.load(f, object_pairs_hook=OrderedDict)
with open(os.path.join(dir_data, fn_input_AOR), "r") as f:
    aor = json.load(f, object_pairs_hook=OrderedDict)
with open(os.path.join(dir_data, fn_input_template), "r") as f:
    res = json.load(f, object_pairs_hook=OrderedDict)

root_l = aol["msg"][0]
root_r = aor["msg"][0]
root_l["parent_structure_id"] = 4000000000 # parent structure ID for root_L
root_r["parent_structure_id"] = 4000000000 # for root_R

target = {'OV_L', 'MY-sat_L', 'EW_L'}
# These IDs are deleted, leaving 'OV_R', 'MY-sat_R', and 'EW_R'
# because these are located at the midline of the brain. 
```

```python
delete_recur(root_l, target)
delete_recur(root_r, target)

res["msg"][0]["children"].append(root_l)
res["msg"][0]["children"].append(root_r)
```

# Save json file

```python
with open(os.path.join(dir_data, fn_output_AO), "w") as f:
    json.dump(res, f, indent=4)
```
